```python
# selecting the relevant features for recommendation
selected_features = df[['Title', 'Description', 'Director', 'Genres', 'Cast', 'Rating', 'Content Type']]
selected_features.set_index('Title', inplace=True)
```

```python
selected_features.head()
```

```python
# replacing the null valuess with null string

# for feature in selected_features:
#   df[feature] = df[feature].fillna('')
# we can't just drop the NaN values because we will lose lots of movies to be given, instead we just fill the NaN values with empty string
selected_features.fillna('', inplace=True)
```

```python
selected_features.isna().sum()
```

```python
import string
def separate(texts):
    t = []
    for text in texts.split(','):
        t.append(text.replace(' ', '').lower())
    return ' '.join(t)

def remove_space(texts):
    return texts.replace(' ', '').lower()

def remove_punc(texts):
    return texts.translate(str.maketrans('','',string.punctuation)).lower()
```

```python
selected_features['Director'] = selected_features['Director'].apply(separate)
selected_features['Genres'] = selected_features['Genres'].apply(separate)
selected_features['Cast'] = selected_features['Cast'].apply(separate)
selected_features['Rating'] = selected_features['Rating'].apply(remove_space)
selected_features['Content Type'] = selected_features['Content Type'].apply(remove_space)
selected_features['Description'] = selected_features['Description'].apply(remove_punc)

selected_features.head()
```

```python
# combining all the selected features

combined_features = df['Description']+' '+df['Director']+' '+df['Genres']+' '+df['Cast']+' '+df['Rating']+' '+df['Content Type']
```

```python
# converting the text data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
selected_features['bag_of_words'] = ''

# Combine all the words into 1 column
for i, row in enumerate(selected_features.iterrows()):
    string = ''
    for col in selected_features.columns:
        if row[1][col] == '':
            continue
        else:
            string += row[1][col] + ' '
            selected_features['bag_of_words'][i] = string.strip()

selected_features.drop(selected_features.columns[:-1], axis=1, inplace=True)
```

```python
selected_features.head(20)
```

```python
tfid = TfidfVectorizer()
tfid_matrix = tfid.fit_transform(selected_features['bag_of_words'])
```

```python
tfid_matrix.shape
```

```python
cosine_sim = cosine_similarity(tfid_matrix, tfid_matrix)
cosine_sim
```

```python
cosine_similarity(tfid_matrix).shape
```

```python
# Later on we will combine with similarity as a column
final_df = df[['Title', 'Content Type']]
```

```python
def recommendation(title, total_result=5, threshold=0.5):
    # Get the index
    idx = final_df[final_df['Title'] == title].index[0]
    # Create a new column for similarity, the value is different for each title you input
    final_df['similarity'] = cosine_sim[idx]
    sort_final_df = final_df.sort_values(by='similarity', ascending=False)[1:total_result+1]

    # You can set a threshold if you want to norrow the result down
    #sort_final_df = sort_final_df[sort_final_df['similarity'] > threshold]

    # Is the title a movie or tv show?
    movies = sort_final_df['Title'][sort_final_df['Content Type'] == 'Movie']
    tv_shows = sort_final_df['Title'][sort_final_df['Content Type'] == 'TV Show']

    if len(movies) != 0:
        print('Similar Movie(s) list:')
        for i, movie in enumerate(movies):
            print('{}. {}'.format(i+1, movie))
        print()
    else:
        print('Similar Movie(s) list:')
        print('-\n')

    if len(tv_shows) != 0:
        print('Similar TV_show(s) list:')
        for i, tv_show in enumerate(tv_shows):
            print('{}. {}'.format(i+1, tv_show))
    else:
        print('Similar TV_show(s) list:')
        print('-')
```
